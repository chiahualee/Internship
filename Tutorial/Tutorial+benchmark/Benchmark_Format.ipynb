```python
import os
rootdir = "/home/chiahual/results"

for directories, subdirs, files in os.walk(rootdir):
    print(directories, subdirs, len(files))
```

```python
temp = ""
for files in os.walk(rootdir):
    temp = files[2]  
```

```python
import pandas as pd
df= pd.DataFrame()
```

```python

for i in range(len(temp)):
    df.loc[i,"resource"]=temp[i][11:]
    with open(rootdir+"/"+temp[i], encoding="latin-1") as f:
        
        for k, line in enumerate(f):
            if k == 38:
                df.loc[i,"naive bayes"]=line[4:8]
            elif k == 40:
                df.loc[i,"KNN"]=line[4:10]
            elif k == 42:
                df.loc[i,"SVM Linear"]=line[4:8]
                
```

```python
df["gpu/core"]=df["resource"].str[-2:]
df["gpu/core"]=df["gpu/core"].str.replace(":","")
df.sort_values(["resource","gpu/core"])
df["resource"]=df["resource"].str[:-2]
df["resource"]=df["resource"].str.replace(":","")

df = df[['resource','gpu/core', 'naive bayes','KNN','SVM Linear']]
```

```python
df["gpu/core"]=df["gpu/core"].astype(int)
```

```python
df.sort_values(["resource","gpu/core"],ascending=True)
```
